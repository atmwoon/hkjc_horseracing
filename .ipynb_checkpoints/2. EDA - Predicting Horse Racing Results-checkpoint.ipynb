### 2.1 Data preparation

#### 2.1.1 Horse Form
Preparing local racing results from data scraping spider `hkjc-horse`

Data transformations performed:
- Convert LBW (length behind winner) from string to numerical associating close distances i.e. NOSE to decimal numbers
- Convert finish time to seconds
- Generate distance groups based on HKJC rules e.g. 'ST Turf 1000'='short'
- Generate course group combining venue, course type and distance group e.g. 'ST Turf short'

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import re
```

```python
import warnings
warnings.simplefilter('ignore')
```

```python
orig_horseform = pd.read_csv("./hkjc/horseform_202008031833.csv")
orig_horseform_rerun = pd.read_csv("./hkjc/horseform_202008041145.csv")
stg_horseform = pd.concat([orig_horseform, orig_horseform_rerun])
```

```python
stg_horseform = stg_horseform[stg_horseform['race_class']!="Race Class"]
```

```python
stg_horseform['placing_no'] = stg_horseform['placing'].apply(lambda x: int(x) if re.search(r'\d+',x) else None)
stg_horseform['venue'] = stg_horseform['course'].apply(lambda x: x.split('/')[0].strip())
stg_horseform['course_type'] = stg_horseform['course'].apply(lambda x: x.split('/')[1].strip(' \"'))
```

```python
lbw_d = {'DH':0, '+NOSE':0.05, 'NOSE':0.1, 'SH':0.15, 'HD':0.2, 'N':0.25, 
         '1/2':0.5, '3/4':0.75, 'ML':99, 'TO':99, '--':None, '---':None}

def lbw_no(placing_no, lbw):
    if lbw in lbw_d:
        lbw_dist = lbw_d[lbw]
    else:
        lbw_list = re.split('-|/', lbw)
        lbw_dist = float(lbw_list[0]) if len(lbw_list)==1 \
                    else float(lbw_list[0]) + float(lbw_list[1])/float(lbw_list[2])
    
    lbw_dist = lbw_dist * -1 if placing_no == 1 else lbw_dist
        
    return lbw_dist

stg_horseform['lbw_no'] = stg_horseform.apply(lambda row: lbw_no(row['placing_no'], row['lbw']), axis=1)
```

```python
stg_horseform['win_odds'] = stg_horseform['win_odds'].apply(lambda x: float(x) if x != '--' else None)
```

```python
def time_seconds(finish_time):
    if finish_time in ['-','--']:
        finish_s = None
    else:
        time_split = finish_time.split('.')
        finish_s = float(time_split[0])*60 + float(time_split[1]) + float(time_split[2])/100
    return finish_s

stg_horseform['finish_time_s'] = stg_horseform['finish_time'].apply(lambda x: time_seconds(x))
```

```python
distance_d = {'ST Turf 1000':'short', 'ST Turf 1200':'short', 'ST Turf 1400':'short', 'ST Turf 1600':'middle',
             'ST Turf 1800':'middle', 'ST Turf 2000':'long', 'ST Turf 2200':'long', 'ST Turf 2400':'long',
             'ST AWT 1200':'short', 'ST AWT 1650':'middle', 'ST AWT 1800':'middle', 'HV Turf 1000':'short',
             'HV Turf 1200':'short', 'HV Turf 1650':'middle', 'HV Turf 1800':'middle', 'HV Turf 2200':'long'}

stg_horseform['distance_group'] = stg_horseform.apply(lambda row: \
                    distance_d[row['venue']+' '+row['course_type']+' '+str(row['distance'])] \
                    if row['venue']+' '+row['course_type']+' '+str(row['distance']) in distance_d \
                    else None, axis=1)
```

```python
stg_horseform['course_group'] = stg_horseform.apply(lambda row: row['venue']+' '+row['course_type']+' '+ \
                                                   (row['distance_group'] if row['distance_group'] else ""), axis=1)
```

```python
stg_horseform['gear'] = stg_horseform.apply(lambda row: row['declared_wt'] if str(row['gear']) == 'nan' \
                                            else row['gear'], axis=1)
stg_horseform['declared_wt'] = stg_horseform['declared_wt'].apply(lambda x: int(x) if re.search(r'^\d+$',x) else None)
```

```python
stg_horseform.head()
```

```python
stg_horseform = stg_horseform.drop_duplicates(subset=['horse_id', 'date'])
```

#### 2.1.3 Overseas
Preparing horse information from pdf parsing overseas URL

Data transformations performed:
- Convert position to placing number
- Convert margin using same logic as LBW in horseform
- Convert odds to numerical format
- Convert race time to seconds

```python
stg_overseas = pd.read_csv("../2. Data/hkjc-overseas-merged.csv")
```

```python
stg_overseas.head()
```

```python
stg_overseas['date'] = stg_overseas['date'].apply(lambda x: pd.to_datetime(x.split('|')[0], format='%d.%m.%Y'))
```

```python
stg_overseas['placing_no'] = stg_overseas['position'].apply(lambda x: int(x.split('/')[0].strip()) \
                                                           if re.search(r'\d+',x.split('/')[0].strip()) else None)
```

```python
stg_overseas['actual_weight'] = stg_overseas['weight_carried'].apply(lambda x: float(x.split('(')[0].strip()))
```

```python
margin_d = {'DH':0, 'H':0, 'Nose':0.1, 'N':0.1, 'SH':0.15, 'SHd':0.15, 'Hd':0.2, 'LHd':0.2, 
            'Hhd':0.2, 'HHd':0.2, 'SNk':0.225, 'Nk':0.25, 'LNk':0.275, 'HNk':0.275, 
            '1/2':0.5, '3/4':0.75, 'ML':99, 'TO':99, '':None, 'N/A':None, 'UR':None}

def margin_no(margin):
    lead_multiplier = -1 if margin[0] == '-' else 1
    if margin[1:] in margin_d:
        margin_dist = margin_d[margin[1:]]
    else:
        margin_dist = float(re.sub(r'[^\d|.]+','',margin))
    
    if margin_dist:
        margin_dist = margin_dist * lead_multiplier
        
    return margin_dist

stg_overseas['margin_no'] = stg_overseas['margin'].apply(lambda x: margin_no(x))
```

```python
odds_d = {'EvensF':1, 'Evs':1, '-':None, 'nan':None}

def odds_parse(odds):
    odds_s = str(odds)
    if odds_s in odds_d:
        new_odds = odds_d[odds_s]
        new_odds = None
    else:
        odds_l = re.sub(r'[^\d|/.]+','',odds_s).split('/')
        new_odds = float(odds_l[0].strip()) if len(odds_l) == 1 \
                    else float(odds_l[0].strip()) / float(odds_l[1].strip()) + 1
        
    return new_odds

stg_overseas['win_odds'] = stg_overseas['odds'].apply(lambda x: odds_parse(x))
```

```python
stg_overseas['race_time_s'] = stg_overseas['race_time'].apply(lambda x: time_seconds(x))
```

```python
stg_overseas = stg_overseas.drop_duplicates(subset=['date', 'overseas_url'])
```

`placing_attributes` function calculates win and place rates on dataset with respect to current race only including races prior:
- df = dataframe to append win & place rates
- prefix = column naming prefix

```python
def placing_attributes(df, prefix):
    df[f'{prefix}_win'] = (df['placing_no'] == 1) * 1
    df[f'{prefix}_place'] = (df['placing_no'] <= 3) * 1
    df = df.sort_values(['horse_id', 'date'])
    df[f'{prefix}_races'] = df.groupby(['horse_id']).cumcount()+1
    df[f'{prefix}_winrate'] = round(df.groupby(['horse_id'])[f'{prefix}_win'].cumsum() / df[f'{prefix}_races'],2)
    df[f'{prefix}_placerate'] = round(df.groupby(['horse_id'])[f'{prefix}_place'].cumsum() / df[f'{prefix}_races'],2)
    return df
```

#### 2.2.2 Form
Form was analysed by looking back at x number of most recent performances across multiple different attributes. 

Moving averages of past performance were calculated using simple moving average (SMA) and exponential weighted averges (EWA).

`moving_average_plot` function generates a heatmap correlation plot of the moving averages for a given data attribute against placing number, win, and place to evaluate predictive power of the feature
- col = data attribute to perform moving average on 
- prefix = prefix of win, place data attributes in dataframe
- df_orig = dataframe to analyse
- sma_window = window range to generate simple moving averages on 
- ewa_span = span range to generate exponential moving average on
- ax_i = matplotlib ax if multiple charts, default = -1 if single chart
- cbar = colourbar formatting for multiple charts
- grouping = data attributes to group by to generate moving average

```python
def moving_average_plot(col, prefix, df_orig, sma_window, ewa_span, ax_i=-1, cbar = True, grouping = None):
    df = df_orig.copy()
    horse_grp = ['horse_id'] + grouping if grouping else ['horse_id']
    
    df.sort_values(horse_grp + ['date'])
    
    df = df.assign(**{f'{col}_sma_{i}': df.groupby(horse_grp)[col] \
                      .transform(lambda x: x.rolling(window=i).mean().round(2)) \
                      for i in sma_window})

    df = df.assign(**{f'{col}_ewa_{i}': df.groupby(horse_grp)[col] \
                       .transform(lambda x: x.ewm(span=i).mean().round(2)) \
                       for i in ewa_span})
    
#     display(df[['horse_id', 'date'] + \
#             list(df.columns[df.columns.str.startswith(col)])].head(10))
    
    next_race = list(df.columns[df.columns.str.startswith(col)])

    df = df.assign(**{f'prev_{col}': df.groupby(horse_grp)[col].shift(1) \
                       for col in next_race})
    
    df_ma = df[['horse_id', 'date', 'placing_no', f'{prefix}win', f'{prefix}place'] + \
                list(df.columns[df.columns.str.startswith('prev_')])]

    corr = df_ma.corr().filter(regex = '[A-Za-z]+$', axis=1)
    
    if ax_i == -1:   
        sns.heatmap(corr, annot = True, cmap='coolwarm', ax = ax, vmin= -1, vmax = 1,) 
    elif cbar:
        sns.heatmap(corr, annot = True, cmap='coolwarm', ax = ax[ax_i],
                   cbar = True, vmin= -1, vmax = 1, yticklabels = False)
    else:
        sns.heatmap(corr, annot = True, cmap='coolwarm', ax = ax[ax_i],
                   cbar = False, vmin= -1, vmax = 1, yticklabels = True)
```

For `placing_no` there is slightly higher correlation between moving averages and win and place when including the overseas races. Exponential weighted means outperform simple moving averages and have benefit of having fewer nulls, simple moving average will always have number of nulls equal to the moving window + 1 for previous race shift.

Best performing moving average is `ewa = 3` and will be included in features for modelling

```python
fig, ax = plt.subplots(figsize=(16, 10), ncols=2)
plot_col = 'placing_no'
moving_average_plot(plot_col, 'local_', df_horseform, range(3, 7), range(1, 11), 0, False)
moving_average_plot(plot_col, '', df_form, range(3, 7), range(1, 11), 1, True)

fig.suptitle(f'EDA Moving Average for {plot_col}')
ax[0].set_title('Local Races')
ax[1].set_title('Local and Overseas Races')

fig.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

#### 2.2.3 Speed
Analysing average speed in previous performances to investigate impact on win / place

`mean_speed` function calculates the average speed for a given horse with respect to current race, only taking prior races into account. Forward filling is used to account for cases where horses are withdrawn or did not participate in a race and their prior mean speed is used
- df = dataframe with input data 
- speed_col = name of speed column in input data
- grouping = data attributes to group by for speed calculation
- new_col = name of new column

```python
def mean_speed(df, speed_col, grouping, new_col):
    df = df.sort_values(grouping + ['date'])
    df[f'{new_col}_counter']= df[speed_col].apply(lambda x: 0 if pd.isnull(x) else 1)
    df[new_col] = round(df.groupby(grouping)[speed_col].cumsum() / \
                            (df.groupby(grouping)[f'{new_col}_counter'].cumsum()), 2)
    df[new_col] = df.groupby(grouping)[new_col].ffill()
    
    df[f'prev_{new_col}'] = df.groupby(grouping)[new_col].shift(1)
    df.drop(columns=[f'{new_col}_counter'], inplace=True)
    return df
```

```python
df_form = mean_speed(df_form, 'speed', ['horse_id'], 'mean_speed_all')
df_form = mean_speed(df_form, 'speed', ['horse_id', 'local'], 'mean_speed_local')
```

`min_max_scale` function performs min max scaling on column

```python
from sklearn.preprocessing import MinMaxScaler

def min_max_scale(df, col, new_col):
    scaler = MinMaxScaler()
    df[[new_col]] = scaler.fit_transform(df[[col]])
    return df
```

```python
df_form = min_max_scale(df_form, 'prev_mean_speed_all', 'prev_mean_speed_all_mm')
df_form = min_max_scale(df_form, 'prev_mean_speed_local', 'prev_mean_speed_local_mm')
```

`rel_speed` function calculates relative speed against the mean speed for all horses in the same grouping and calculating a difference with the average of all winners, placers and all racers. This difference then has a moving average applied to it to see how it changes over time based on horse form / performance
- grouping = group by to calculate speed on 
- new_col = name of new column
- sma_window = window range for simple moving average
- ewa_span = span range for exponential weighted average

```python
def rel_speed(grouping, new_col, sma_window, ewa_span):
    df = df_speed.copy()
    horse_grp = ['horse_id'] + grouping
    
    df = df.sort_values(horse_grp + ['date'])
    df[f'{new_col}_counter']= df['speed'].apply(lambda x: 0 if pd.isnull(x) else 1)
    df[new_col] = round(df.groupby(horse_grp)['speed'].cumsum() / \
                            (df.groupby(horse_grp)[f'{new_col}_counter'].cumsum()), 2)
    df[new_col] = df.groupby(horse_grp)[new_col].ffill()
    
    df = df.merge(df[df['local_win']==1].groupby(grouping + ['date'])['speed'] \
                                 .apply(lambda x: round(x.mean(),2)) \
                                 .reset_index().rename(columns={'speed':f'{new_col}_win'}),
               on = grouping + ['date'], how = 'left')
    
    df = df.merge(df[df['local_place']==1].groupby(grouping + ['date'])['speed'] \
                                 .apply(lambda x: round(x.mean(),2)) \
                                 .reset_index().rename(columns={'speed':f'{new_col}_place'}),
               on = grouping + ['date'], how = 'left')
    
    df[f'{new_col}_avg'] = df.groupby(grouping + ['date'])['speed'].transform(lambda x: round(x.mean(),2))
    
    df[f'{new_col}_place'] = df.apply(lambda row: row[f'{new_col}_avg'] if pd.isnull(row[f'{new_col}_place']) \
                                      else row[f'{new_col}_place'],axis=1)
    df[f'{new_col}_win'] = df.apply(lambda row: row[f'{new_col}_place'] if pd.isnull(row[f'{new_col}_win']) \
                                      else row[f'{new_col}_win'],axis=1)
    
    placing = ['win', 'place', 'avg']
    
    df = df.assign(**{f'rel_{i}_diff': df['speed'] - df[f'{new_col}_{i}'] for i in placing})
    
    for col in list(df.columns[df.columns.str.startswith('rel_')]):
        df = df.assign(**{f'{col}_sma_{i}': df.groupby(horse_grp)[col] \
                           .transform(lambda x: x.rolling(window=i).mean().round(2))  \
                           for i in sma_window})

        df = df.assign(**{f'{col}_ewa_{i}': df.groupby(horse_grp)[col] \
                           .transform(lambda x: x.ewm(span=i).mean().round(2)) \
                           for i in ewa_span})
    
    next_race = list(df.columns[df.columns.str.startswith('rel_')])

    df = df.assign(**{f'prev_{col}': df.groupby(['horse_id'])[col].shift(1) \
                       for col in next_race})
    
    df.drop(columns=[f'{new_col}_counter'], inplace=True)
    
    corr_dict = {}

    cols = ['placing_no', 'local_win', 'local_place', 'speed']

    for p in placing:
        rows = cols + list(df.columns[df.columns.str.startswith(f'prev_rel_{p}')])
        corr_dict[p] = df.corr().filter(items=cols, axis=1).filter(items = rows,axis=0)
    
    fig, ax = plt.subplots(figsize = (18,10), ncols=3)

    sns.heatmap(corr_dict['win'], annot = True, cmap='coolwarm', ax = ax[0],
                   cbar = False, vmin= -1, vmax = 1, yticklabels = True)
    sns.heatmap(corr_dict['place'], annot = True, cmap='coolwarm', ax = ax[1],
                   cbar = False, vmin= -1, vmax = 1, yticklabels = False)
    sns.heatmap(corr_dict['avg'], annot = True, cmap='coolwarm', ax = ax[2],
                   cbar = True, vmin= -1, vmax = 1, yticklabels = False)

    plt.suptitle('Relative Speed Moving Average for {0}'.format(' '.join(grouping)))
    ax[0].set_title('Win')
    ax[1].set_title('Place')
    ax[2].set_title('Average')

    plt.show()
    
    return df
```

#### 2.2.6 Winrate Groupings 
Investigate winrate by different groupings

`group_placing` function calculates winrate by defined grouping:
- df = dataframe of input data
- prefix = naming prefix
- grouping = data attributes to group by

```python
def group_placing(df, prefix, grouping=None):
    horse_grp = ['horse_id'] + grouping if grouping else ['horse_id']
    df = df.sort_values(horse_grp + ['date'])
    df[f'{prefix}_races'] = df.groupby(horse_grp).cumcount()+1
    df[f'{prefix}_winrate'] = round(df.groupby(horse_grp)['local_win'].cumsum() / df[f'{prefix}_races'],2)
    df[f'{prefix}_placerate'] = round(df.groupby(horse_grp)['local_place'].cumsum() / df[f'{prefix}_races'],2)
    df[f'prev_{prefix}_winrate'] = df.groupby(horse_grp)[f'{prefix}_winrate'].shift(1).fillna(0)
    df[f'prev_{prefix}_placerate'] = df.groupby(horse_grp)[f'{prefix}_placerate'].shift(1).fillna(0)
    
    return df
```

```python
df_placing = df_prev_diff = df_horseform[df_horseform['venue'].isin(['ST','HV'])] \
                            [['horse_id', 'horse', 'date','course', 'distance', 'going', 
                              'placing_no','course_type', 'distance_group', 
                             'course_group', 'local_win', 'local_place']]

groupings = [['distance'], ['going'], ['course_type'], ['distance_group'], 
             ['course_type', 'distance_group'], ['course_group']]

for grouping in groupings:
    df_placing = group_placing(df_placing, '_'.join(grouping), grouping)
    
df_placing = group_placing(df_placing, 'all')
```

```python
df_draw['draw_winrate'] = round(df_draw.groupby(['course_group','draw'])['draw_win'].cumsum() \
                                  / df_draw.groupby(['course_group','draw'])['draw_races'].cumsum(),2).fillna(0)
df_draw['draw_placerate'] = round(df_draw.groupby(['course_group','draw'])['draw_place'].cumsum() \
                                  / df_draw.groupby(['course_group','draw'])['draw_races'].cumsum(),2).fillna(0)

min_max_tr = lambda x: (x - x.min()) / (x.max() - x.min())

df_draw['draw_winrate_scaled'] = df_draw.groupby(['course_group', 'date'])['draw_winrate'].transform(min_max_tr)
df_draw['draw_placerate_scaled'] = df_draw.groupby(['course_group', 'date'])['draw_placerate'].transform(min_max_tr)

df_draw = df_draw.assign(**{f'prev_{col}': df_draw.groupby(['course_group','draw'])[col].shift(1).fillna(0) \
                       for col in list(df_draw.columns[df_draw.columns.str.endswith(('rate','scaled'))])})
```

```python
df_draw_placing = pd.merge(df_horseform[(df_horseform['venue'].isin(['ST','HV'])) & (df_horseform['date']>'2016-09-01')], 
                           df_draw, on = ['course_group', 'draw', 'date'], how='left')
```

Calculated win and place rates grouped by course and draw do not show good correlation with win and place, draw by itself shows better correlation and will be included in the features for modelling

```python
df_draw_placing = df_draw_placing[['horse_id','horse', 'placing_no', 'local_win', 'local_place', 'draw',
                                  'prev_draw_winrate', 'prev_draw_placerate',
                                  'prev_draw_winrate_scaled', 'prev_draw_placerate_scaled']]

fig, ax = plt.subplots(figsize=(8,10))
corr_plot = df_draw_placing.corr()

sns.heatmap(corr_plot, annot = True, cmap='coolwarm', ax = ax, vmin= -1, vmax = 1,) 
plt.show()
```
